### Fill Null values by central tendency (Mean, Mode, Median)
We can fill the null values with different approaches depending on the type of data we get. For categorical data, it is preferred to use Mode while for numerical values we will be utilizing mean and median. Here we have filled the null values with mode for the categorical columns ‘club’ and ‘joined’.

```python
cols = ["Joined", "Club"]
df[cols] = df[cols].fillna(df.mode().iloc[0])
```
