```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plot

```python
x = [1, 4, 6, 3, 9, 3, 5, 6, 7, 4, 3, 5]  # values in x-column
y = [2, 4, 6, 7, 4, 2, 4, 6, 1, 2, 2, 1]  # values in y-column
```

```python
plt.figure()
plt.scatter(x, y, c='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Barchart

```python

data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())  # extracting only the keys from data
values = list(data.values())  # extracting only the values from data
fig = plt.figure(figsize = (10, 5))

# creating the bar chart
plt.bar(courses, values, color ='maroon', width = 0.4)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

# Line Chart

```python
linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8])
quadratic_data = linear_data**2
plt.figure()
plt.plot(linear_data, '-o', quadratic_data, '-o')
plt.legend(['Linear', 'Quadratic'])
plt.show()
```
